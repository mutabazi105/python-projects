## 1. Lists

A **list** is an ordered, mutable collection of elements.

Typical use cases:
- Collections of items that may change
- Sequential data
- Accumulation of results

```python
numbers = [10, 20, 30, 40]
names = ["Ada", "Alan", "Grace"]

print(numbers)
print(names)
```

We can use `range()` method to create a list of integers

```python
numbers = list(range(10, 101, 10))
print(numbers)
```

### Indexing and slicing (lists)

- Indexing starts at `0`
- Negative indices count from the end
- Slicing extracts a sub-list

```python
values = [0, 1, 2, 3, 4, 5]

print(values[0])      # first element
print(values[-1])     # last element
print(values[1:4])    # slice from index 1 to 3
print(values[:3])     # first three elements
print(values[3:])     # from index 3 to end
```

### Modifying lists

Lists are **mutable**, meaning they can be changed after creation.

```python
items = ["apple", "banana"]
items.append("orange")
items.append("banana")

items.remove("banana")
items.pop(-2)
items.insert(-1, 'AAA')
items.remove("banana")

items[0] = items[0].upper()

items
```

### Useful list methods

Python lists have several built-in methods that are very handy.
- `append()`: Adds an element to the end.
- `insert()`: Adds an element at a specific index.
- `pop()`: Removes and returns an element (default is the last one).
- `remove()`: Removes the first occurrence of a value.
- `sort()`: Sorts the list in place.
- `reverse()`: Reverses the list in place.
- `extend()`: Extend list by appending multiple elements from another list

```python
colors = ["red", "blue", "green", "yellow"]

# Sorting and reversing
colors.sort()
print("Sorted:", colors)

colors.reverse()
print("Reversed:", colors)

# Removing items
last = colors.pop()
print("Popped:", last)
print("Remaining:", colors)
```

### List Comprehensions

List comprehensions provide a concise way to create lists. They are often faster and more readable than standard loops for creating new lists.

Syntax: `[expression for item in iterable if condition]`

```python
numbers = [1, 2, 3, 4, 5]

for i in list(range(len(numbers))):
    if numbers[i] % 2 == 0:
        print(f"{numbers[i]} is even")
```

```python
numbers = [1, 2, 3, 4, 5]

# Creating a list of squares using a loop
squares_loop = []
for n in numbers:
    squares_loop.append(n ** 2)

# Creating a list of squares using comprehension
squares_comp = [n ** 2 for n in numbers]

print(squares_loop)
print(squares_comp)

# With a condition (even numbers only)
evens = [n**2 for n in numbers if n % 2 == 0]
print("Evens:", evens)
```

## 2. Tuples

A **tuple** is an ordered, **immutable** collection.

Typical use cases:
- Fixed collections of values (e.g., coordinates `(x, y)`).
- Records where structure matters more than order.
- Faster and safer than lists for read-only data.


```python
point = (10, 20)
person = ("Ada", "Lovelace", 36)

print(point)
print(person)
```

```python
a = point[0]
b = point[1]
print(f"x: {a}, y: {b}")
```

```python
a, b = point
print(f"x: {a}, y: {b}")
```

### Tuple unpacking

```python
x, y = point
first, last, age = person

print(x, y)
print(first, last, age)
```

## 3. Dictionaries

A **dictionary** is an unordered collection that stores **key–value pairs**. It provides fast lookups based on keys.

Typical use cases:
- Mapping unique IDs to data (e.g., User ID → User Profile).
- Structured records (e.g., representing a JSON object).
- Counting/Frequency tables.

```python
contact = {
    "name": "Ada Lovelace",
    "email": "ada@example.com",
    "age": 36,
    42: "The answer"
}
```

```python
contact = {
    "name": "Ada Lovelace",
    "email": "ada@example.com",
    "age": 36,
    42: "The answer"
}

print(contact)
print(contact["name"])
```

### Adding, updating, and removing entries

```python
contact["phone"] = "+44-1234"
contact["age"] = 37
del contact["email"]

print(contact)
```

### Safely accessing data with `get()`

Accessing a missing key like `contact["address"]` raises a `KeyError`.
Using `.get()` returns `None` (or a default value) instead.

```python
# This is safe even if "address" doesn't exist
address = contact.get("address", "Unknown Address")
print("Address:", address)

# Traditional check
if "address" in contact:
    print(contact["address"])
else:
    print("No address found")
```

```python
for key, value in contact.items():
    print(f"{key}: {value}")
```

### Iterating over dictionaries

```python
for key in contact:
    print(key, "->", contact[key])

print("---")

for key, value in contact.items():
    print(key, "->", value)
```

## 4. Sets

A **set** is an unordered collection of **unique** elements. It mimics mathematical sets.

Typical use cases:
- Removing duplicates from a list.
- Membership testing (is *x* in *y*?) – much faster than lists!
- Set operations (Union, Intersection, Difference).

```python
# Menge

numbers = [1, 2, 2, 3, 3, 4, 4, 4, 7]
unique_numbers = set(numbers)

print(unique_numbers)
```

### Set operations

```python
a = set([1, 2, 3])
b = {3, 4, 5}

print(a | b)    # union
print(a & b)    # intersection
print(a - b)    # difference
print(b - a)    # difference

print(type(a))
```

## 5. Advanced: Nested Data Structures

Real-world data is often complex. You can nest data structures within each other to model more sophisticated information.

- **List of dictionaries**: Common for rows of data (like a database query result or CSV).
- **Dictionary of lists**: Grouping items by category.

```python
users = [
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Bob"}
]

users[1]['name']

print(type(users))
print(type(users[0]))
print(type(users[0]['name']))
```

```python
# List of dictionaries (like a database)
users = [
    {"id": 1, "name": "Alice"},
    {"id": 2, "name": "Bob"}
]

print(users[0]["name"])

# Dictionary of lists (e.g., student grades)
grades = {
    "Alice": [85, 90, 88],
    "Bob": [78, 82, 80]
}

print("Alice's first grade:", grades["Alice"][0])
```

```python
a = [1, 2, 3]

b = a.copy()

b[0] += 1

print(a)
print(b)
```

## 6. Advanced: Reference vs. Copy

A common pitfall in Python is assuming assignment creates a copy.
- Assigning a list to a new variable creates a **reference** (alias). Changes to one affect both!
- To create a true independent duplicate, use the `.copy()` method or slicing `[:]`.

```python
a = [1, 2, 3]
b = a          # Reference (alias)
c = a.copy()   # Independent copy

b.append(4)    # Modifies 'a' as well!
c.append(5)    # Modifies only 'c'

print("a:", a)
print("b:", b)
print("c:", c)
```

## 7. Summary: Iteration Patterns

Iteration syntax depends on the structure, but `for ... in ...` is the universal pattern. This section consolidates how to loop over each type.

```python
# List iteration
for n in [1, 2, 3]:
    print("list item:", n)

# Tuple iteration
for item in ("a", "b", "c"):
    print("tuple item:", item)

# Dictionary iteration
data = {"a": 1, "b": 2}
for key, value in data.items():
    print("dict:", key, value)

# Set iteration
for x in {10, 20, 30}:
    print("set item:", x)
```

## 8. Strategy: Choosing the Right Data Structure

Selecting the appropriate structure is critical for performance and code clarity.

| Requirement | Best Choice | Why? |
| :--- | :--- | :--- |
| **Ordered**, changeable sequence | **List** | Flexible, easy to modify. |
| **Fixed**, immutable sequence | **Tuple** | Safer, memory efficient, hashable. |
| **Key → Value** lookup | **Dictionary** | O(1) lookup speed, structured data. |
| **Unique** items / membership test | **Set** | Auto-deduplication, fast checking. |
| **Complex/Hierarchical** data | **Nested Structures** | Models real-world objects/tables. |

## Exercise 2 — Word frequency analysis

Given a text, compute how often each word appears.

Steps:
1. Split text into words
2. Use a dictionary to count occurrences
3. Print the result

```python
# Exercise 2 solution cell

text = "python is great and python is easy to learn"
words = text.split()

freq = {}

for word in words:
    if word not in freq:
        freq[word] = 1
    else:
        freq[word] += 1

for word, count in freq.items():
    print(word, "->", count)
```

## Exercise 3 — Simple data aggregation (without Pandas)

Given a list of numbers:
- Count how many values are greater than 10
- Compute the sum of all values
- Compute the average

```python
# Exercise 3 solution cell

values = [4, 12, 7, 19, 3, 25, 10]

count_gt_10 = 0
total = 0

for v in values:
    total += v
    if v > 10:
        count_gt_10 += 1

average = total / len(values)

print("count_gt_10:", count_gt_10)
print("sum:", total)
print("average:", average)
```
